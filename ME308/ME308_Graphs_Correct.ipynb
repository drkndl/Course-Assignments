```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BSFC = [0.3583, 0.3086, 0.2321, 0.2041]

BP = [0, 1.1666, 2.31666, 3.45, 4.30666]

plt.plot(BP[1:], BSFC, marker='o')
plt.title("Characteristic Curve: BSFC vs BP")
plt.xlabel("Brake Power (kW)")
plt.ylabel("BSFC (kg/kW-hr)")
plt.grid()
plt.ylim(0,0.4)
plt.xlim(0, 4.5)
plt.savefig("graph1")

plt.show()
```

```python
mech_eff = [0, 49.2957, 65.876, 74.193, 78.208]

plt.plot(BP, mech_eff, marker='o')
plt.title("Mechanical Efficiency vs BP")
plt.xlabel("Brake Power (kW)")
plt.ylabel("Mechanical Efficiency (%)")
plt.grid()
plt.savefig("graph3")

plt.show()
```

```python
i_eff = [25.881, 46.357, 40.276, 47.532, 51.279]

plt.plot(BP, i_eff, marker='o')
plt.title("Indicated Thermal Efficiency vs BP")
plt.xlabel("Brake Power (kW)")
plt.ylabel("Indicated Thermal Efficiency (%)")
plt.grid()
plt.ylim(0, 70)
plt.savefig("graph4")

plt.show()
```

```python
b_eff = [0, 22.852, 26.532, 35.266, 40.104]

plt.plot(BP, b_eff, marker='o')
plt.title("Brake Thermal Efficiency vs BP")
plt.xlabel("Brake Power (kW)")
plt.ylabel("Brake Thermal Efficiency (%)")
plt.grid()
plt.savefig("graph5")

plt.show()
```
