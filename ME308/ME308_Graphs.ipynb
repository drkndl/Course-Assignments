```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BSFC = [0.53185287, 0.37418745, 0.322990085, 0.340713131]

BP = [0,0.887881309,
1.739940317,
2.533148402,
3.326356488]

plt.plot(BP[1:], BSFC, marker='x', color='red')
plt.title("BSFC vs BP")
plt.xlabel("Brake Power (kW)")
plt.ylabel("BSFC (kg/kW-hr)")

plt.savefig("jaigraph1")

plt.show()
```
