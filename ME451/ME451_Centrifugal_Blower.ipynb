```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 10                                     # Number of pulses
t = np.array([265, 217, 170, 144, 123])    # Time
h1 = np.array([19, 16, 12, 7.2, 6.1])      # Delivery pressure cm of H2O
h = np.array([0.3, 0.9, 3.2, 5.3, 6.4])    # Head across flow meter
```

```python
motor_eff = 0.9     # Motor efficiency
rho_a = 1.275       # Air density kg/m^3
rho_w = 1000        # Water density kg/m^3
g = 9.81            # Acceleration due to gravity m/s^2
```

```python
# Input power to motor (kW)
IP = 3600*w/(320*t)
IP
```

```python
# Input to blower (kW)
SP = motor_eff * IP
SP
```

```python
# Output to blower (kW)

gamma_a = rho_a*g              # Specific weight of air
Q = 171*np.sqrt(h/rho_a)       # Volume of air delivered m^3/hr
print("Q = ", Q)
H = h1*rho_w / (100*rho_a)     # Head in metres of air
print("H = ", H)
OP = gamma_a*Q*H/(3600*10**3)  # Output of the blower
print("OP = ", OP)
```

```python
# Blower efficiency
blower_eff = OP/SP*100
blower_eff
```

```python
# Plots

plt.plot(h1, blower_eff, marker='o')
plt.grid()
plt.title(r"$\eta_{blower}$ vs $h_1$")
plt.xlabel(r"Delivery head $h_1$ (cm of $H_2O$)")
plt.ylabel(r"Blower efficiency $\eta_{blower}$ (%)")
plt.savefig("Desktop/exp2plot1.png")
plt.show()
```

```python
plt.plot(h1, Q, marker='o', color='orange')
plt.grid()
plt.title(r"Q vs $h_1$")
plt.xlabel(r"Delivery head $h_1$ (cm of $H_2O$)")
plt.ylabel(r"Discharge volume Q ($m^3/hr$)")
plt.savefig("Desktop/exp2plot2.png")
plt.show()
```

```python
plt.plot(h1, SP, marker='o', color='green')
plt.grid()
plt.title(r"SP vs $h_1$")
plt.xlabel(r"Delivery head $h_1$ (cm of $H_2O$)")
plt.ylabel(r"Input to blower SP (kW)")
plt.savefig("Desktop/exp2plot3.png")
plt.show()
```
