```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list1 = [500, 750, 1000, 2000, 4000, 3750, 5000, 6000, 4500, 2000, 1500, 700]
list2 = [250, 750, 1500, 2000, 4500, 3650, 5700, 5500, 3650, 2200, 1000, 640]
plt.plot(range(len(list1)), list1, label="Last Year's Sales")
plt.plot(range(len(list1)), list2, label="This Year's Projections")
plt.title("Seasonal Product Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()

plt.savefig("seasonal.eps", format='eps')

plt.show()
```
