```python
from __future__ import division
from tkinter import *
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
class MyWindow:
    def __init__(self, win):
        self.var = IntVar()
        
        self.lbl1=Label(win, text='End Mass (kg)')
        self.lbl2=Label(win, text='Material')
        self.lbl3=Label(win, text='Damping ratio')
        self.lbl4=Label(win, text='Length (m)')
        self.lbl5=Label(win, text='Cross Section')
        
        self.t1=Entry()
        
        self.t2.set("Steel")
        self.t2 = OptionMenu(win, self.t2, "Steel", "Cast Iron", "some shit")
        self.t2.pack()
        
        self.t3=Entry()
        
        self.t4=Entry()
#         self.t41=Radiobutton(win, text="Free", variable=self.var, value="1", command='sel')
#         self.t42=Radiobutton(win, text="Forced", variable=self.var, value="2", command='sel')

        self.t5=OptionMenu(win, self.t5, "Square", "Circle", "Random")
    
        self.lbl1.place(x=50, y=50)
        self.t1.place(x=250, y=50)
        self.lbl2.place(x=50, y=100)
        self.t2.place(x=250, y=100)
        self.b1=Button(win, text='Calculate', command=self.calc)
        self.b1.place(x=190, y=300)
        self.lbl3.place(x=50, y=150)
        self.t3.place(x=250, y=150)
        self.lbl4.place(x=50, y=200)
        self.t4.place(x=250, y=200)
        self.lbl5.place(x=50, y=250)
        self.t5.place(x=250, y=250)

    def calc(self):
        m=float(self.t1.get())
        k=float(self.t2.get())
        c=float(self.t3.get())
        ohm=float(self.t5.get())

        newWin = Tk()
        newWin.title('A09-Outputs')
        newWin.geometry('500x500')

        wn = (k/m)**(1/2)
        fn = (1/(2*math.pi))*(wn)
        t = 1/fn
        cc = 2*(m*k)**(1/2)
        df = c*cc
        wd = wn*((1-(c**2))**(1/2))
        q = 1/(2*c)
        if str(self.var.get())=="2":
            tr = ((1+((2*c*ohm)/wn)**2)/((1-(ohm/wn)**2)**2+(2*c*ohm/(wn))**2))**1/2
        fr = ohm/fn

        lbl1 = Label(newWin, text='Natural angular frequency (wn)[rad/s]')
        lbl2 = Label(newWin, text='Natural frequency (fn) [Hz]')
        lbl3 = Label(newWin, text='Period of oscillations (T) [s]')
        lbl4 = Label(newWin, text='Critical damping (cc)')
        lbl5 = Label(newWin, text='Damping factor (c)')
        lbl6 = Label(newWin, text='Damped natural frequency (wd)')
        lbl7 = Label(newWin, text='Quality factor (Q)')
        if str(self.var.get())=="2":
            lbl8 = Label(newWin, text='Transmissibility (TR)')

        a1 = Label(newWin, text=str(wn))
        a2 = Label(newWin, text=str(fn))
        a3 = Label(newWin, text=str(t))
        a4 = Label(newWin, text=str(cc))
        a5 = Label(newWin, text=str(df))
        a6 = Label(newWin, text=str(wd))
        a7 = Label(newWin, text=str(q))
        if str(self.var.get())=="2":
            a8 = Label(newWin, text=str(tr))

        lbl1.place(x=50, y=50)
        lbl2.place(x=50, y=100)
        lbl3.place(x=50, y=150)
        lbl4.place(x=50, y=200)
        lbl5.place(x=50, y=250)
        lbl6.place(x=50, y=300)
        lbl7.place(x=50, y=350)
        if str(self.var.get())=="2":
            lbl8.place(x=50, y=400)

        a1.place(x=350, y=50)
        a2.place(x=350, y=100)
        a3.place(x=350, y=150)
        a4.place(x=350, y=200)
        a5.place(x=350, y=250)
        a6.place(x=350, y=300)
        a7.place(x=350, y=350)
        if str(self.var.get())=="2":
            a8.place(x=350, y=400)

        if str(self.var.get())=="2":
            fig, ax = plt.subplots(constrained_layout = True) 
            x = np.arange(0.1, 10, 0.01) 
            z = x/(2*math.pi)
            y = ((1+(2*c*z)**2)/(((1-z**2)**2+(2*c*z)**2))**1/2)
            ax.loglog(x, y) 
            ax.set_xlabel('Frequency Ratio (ohm/fn)') 
            ax.set_ylabel('Transimissibility') 
            ax.set_title('Transmissiblity vs Frequency Ratio Graph(log-log)') 
            plt.show()

window=Tk()
mywin=MyWindow(window)
window.title('Free Vibration on Cantilever Beam - Inputs')
window.geometry("430x500")
window.mainloop()
```
